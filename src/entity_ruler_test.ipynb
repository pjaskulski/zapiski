```python
import spacy
```

```python
from spacy import displacy
```

```python
colors = {"persName": "#feb236", "date":"#bdcebe", "placeName":"#dac292"}
options = {"ents": ["persName", "date", "placeName"], "colors": colors}
```

```python
nlp = spacy.load('pl_core_news_md')
```

```python
with open('../data/bodniak_baltyk_small.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
doc = nlp(text)
```

```python
# Bez użycia komponentu EntityRuler
displacy.render(doc, style='ent', options=options)
```

```python
nlp = spacy.load('pl_core_news_md')

barnim = nlp.add_pipe("entity_ruler")
pattern = [{"label":"persName", "pattern": "Barnim XI"}]
barnim.add_patterns(pattern)
```

```python
doc = nlp(text)
```

```python
# z użyciem EntityRuler
displacy.render(doc, style='ent', options=options)
```

```python
nlp = spacy.load('pl_core_news_md')

wzorce_encji = nlp.add_pipe("entity_ruler")
pattern = [{"label":"placeName", "pattern": "hetman"}]
wzorce_encji.add_patterns(pattern)
doc = nlp(text)
# z użyciem EntityRuler
displacy.render(doc, style='ent', options=options)
```

```python
nlp = spacy.load('pl_core_news_md')

colors = {"persName": "#feb236", "date":"#bdcebe", "placeName":"#dac292", "OCCUPATION":"#ffff00"}
options = {"ents": ["persName", "date", "placeName", "OCCUPATION"], "colors": colors}

wzorce_encji = nlp.add_pipe("entity_ruler")
pattern = [{"label":"OCCUPATION", "pattern": [{"LEMMA": {"IN":["hetman","senator","marszałek", "król"]}}]}]
wzorce_encji.add_patterns(pattern)
doc = nlp(text)
displacy.render(doc, style='ent', options=options)
```
